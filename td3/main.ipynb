```python
# Centrage et réduction des valeurs quantitatives
from sklearn.preprocessing import StandardScaler

std = StandardScaler()
ZTrain = std.fit_transform(df2)
ZTrain
```

```python
print(type(ZTrain))
std_df2 = pd.DataFrame(ZTrain, index=df.index, columns=['puissance', 'cylindree', 'vitesse', 'longueur', 'largeur', 'hauteur',
       'poids', 'CO2'])
std_df2
```

```python
#vérification de l'uniformité de l'ecart-type
import numpy as np

np.std(std_df2)
```

```python
#initialisation de cluster
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, n_init=10, random_state=0).fit(std_df2)
kmeans
```

```python
#inertie intra class
WSS = kmeans.inertia_
WSS
```

```python
#groupe d'appartenance
GA = kmeans.labels_
GA
```

```python
#Attribution de chacun des modèles aux groupes d'appartenance
std_df2["labels"] = GA
std_df2.head()
```

```python
pd.DataFrame(GA, index=df2.index)
```

```python
#Nombre de véhicules par cluster
import numpy as np

np.unique(GA, return_counts=True)
```

```python
#moyennes des variables conditionnellement aux classes
prot_cls = pd.DataFrame(kmeans.cluster_centers_, columns=df.columns[:8])
prot_cls
```

```python
"""
le premier groupe est caractérisée par leur cylindée, le second par leur vistesse et le troisième par leu hauteur
"""
kmeans.cluster_centers_[0]
```

```python
kmeans.cluster_centers_
```

```python
#calcul de l'inertie inter-class BSS
"""
La formule BSS calcule la somme totale pondérée des distances moyennes entre les points de données et leurs centroïdes de cluster. 
Une BSS plus faible indique un meilleur clustering. 
La BSS peut être utilisée pour comparer différents modèles k-means et identifier le nombre optimal de clusters.
"""

effectifs = np.unique(GA, return_counts=True)[1]
BSS = np.sum(effectifs * np.sum(kmeans.cluster_centers_**2, axis=1))
BSS
```

```python
TSS = BSS + WSS
TSS
```

```python
#calcul direct du TSS sachant que TSS = n (nombre total d’individus) * somme des variances
df2.shape[0] * df2.shape[1]
```

```python
#calcul du carré d'inertie
BSS/TSS *100
```

```python
#Courbe d'évolution de l'inertie intra class wss fonction du nombre de classe

#initialisation des valeurs de k
k_value = np.zeros(11)

# Stocker les valeurs d'inertie intra-classes
k_value[1] = TSS

#Itération des valeurs de k
for k in range(2, 11):
    km = KMeans(n_clusters=k, n_init=10, random_state=0).fit(std_df2)
    k_value[k] = km.inertia_
k_value
```

```python
import matplotlib.pyplot as plt
plt.title("WSS")
plt.xlabel("# of clusters")
plt.plot(np.arange(1,11),k_value[1:],marker='X',linestyle='solid')
plt.show()
```

```python
#identifier les variables qui impactent le plus la séparabilité
#calcul des carrés des rapports de corrélation

#SCT = n * la variance
# etpuisque les vaiables sont centrées et réduites
SCT = std_df2.shape[0]

#SCE de chaque variable, sachant qu'elles sont centrées
SCE = np.dot(effectifs,kmeans.cluster_centers_**2)

#carré du rapport de corrélation
eta2 = SCE / SCT

#transformer en series pour une présentation plus avenante
eta2S = pd.Series(eta2)
eta2S.index = std_df2.columns[:8]
print(eta2S.sort_values(ascending=False))
```

```python
#calcul du prix moyen de véhicule pour chaque classe
mean_price = Sdf.pivot_table(index=["labels"], values=["prix"], aggfunc= pd.Series.mean)
mean_price
```

```python
#Calcul du v de cramer
from scipy.stats import chi2_contingency

var_qual = ['origine', 'carburant', 'type4X4']

for var in var_qual:
    crosstab = np.array(pd.crosstab(Sdf['labels'],Sdf[var],rownames=None,colnames=None))
    stat = chi2_contingency(crosstab)[0]
    obs = np.sum(crosstab)
    mini = min(crosstab.shape)-1
    v = np.sqrt(stat/(obs*mini))
    print("v de Cramer de {} = {}".format(var,v))
```

```python
#proportion de type4X4 = oui dans chaque cluster
pd.crosstab(Sdf['labels'],Sdf['type4X4'],normalize='index')
```
